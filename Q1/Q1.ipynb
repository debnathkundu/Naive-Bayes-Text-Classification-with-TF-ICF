##**Computing IDF Vector**

**Collecting all file data in a list for efficient computation:**

```python
All_File_Data=[]
for file_name_ in tqdm(list_dir):
  file_name= file_path+file_name_

  with open(file_name, 'r') as f: #open the file in read mode
      contents = f.read() #reading the file
      All_File_Data.append(contents)

len(All_File_Data)
```

```python
n_docs = len(list_dir)         #Number of documents in the corpus
n_words_set = len(vocab_list) #Number of unique words in the
 
IDF_List = pd.DataFrame(np.zeros((1, n_words_set)), columns=vocab_list)

i=0
for w in tqdm(vocab_list):
    k = 0    # number of documents in the corpus that contain this word

    for i in range(0,len(All_File_Data)):
        #list containing each term in the file as an element
        words=list(All_File_Data[i].split(' '))

        if w in words:
          k += 1
    IDF_List[w][0] =  math.log10(n_docs / k)

    # print(k)
    # print("IDF of: ",f'{w:>15}: {IDF_List[w][0]:>10}' )
```

##(iii) Term frequency : f(t,d) / (t, d)


```python
n_docs = len(list_dir)         #Number of documents in the corpus
n_words_set = len(vocab_list) #Number of unique words in the
 
Term_Frequency_Fraction = pd.DataFrame(np.zeros((n_docs, n_words_set)), columns=vocab_list)

#iterating through each file in directory
i=0
for file_name_ in tqdm(list_dir):
  file_name= file_path+file_name_
  # print(file_name)
  with open(file_name, 'r') as f: #open the file in read mode

      #reading the file
      contents = f.read()
      
      #list containing each term in the file as an element
      words=list(contents.split(' '))
      # print(len(words)," & ",Term_Frequency_Raw_Count.loc[[i]].sum(axis=1) )
      # print()
      # break
      for w in words:
        Term_Frequency_Fraction[w][i] = ( Term_Frequency_Raw_Count[w][i] / (Term_Frequency_Raw_Count.loc[[i]].sum(axis=1)) )

  i=i+1
```

```python
#pd.set_option('display.max_columns', None)
Term_Frequency_Fraction
```

###Analysis
**Pros:** 
1. Term weighting indicates how important each individual word is to the document and within the document collection.
2.  Can be used further to assign weights to different sections of the document such as Title/Body.
3. It represents the relative frequency of term t within document d,
**Cons:** 
1. Still no information of positional index. The method does not capture the semantic correlation between different terms in a document.
2. No information about rare terms.
3. Relevance of a document does not increase proportionally with term count.
4. Raw Count isnt the best representation of the text because common words such as "the", "a", "to" are almost always the terms with highest frequency in the text.

##(iv) Log normalization log(1+f(t,d))


```python
n_docs = len(list_dir)         #Number of documents in the corpus
n_words_set = len(vocab_list) #Number of unique words in the
 
Term_Frequency_Log_Normalisation = pd.DataFrame(np.zeros((n_docs, n_words_set)), columns=vocab_list)
 
#iterating through each file in directory
i=0
for file_name_ in tqdm(list_dir):
  file_name= file_path+file_name_

  with open(file_name, 'r') as f: #open the file in read mode

      #reading the file
      contents = f.read()
      
      #list containing each term in the file as an element
      words=list(contents.split(' '))
      for w in words:
        Term_Frequency_Log_Normalisation[w][i] = math.log10(Term_Frequency_Raw_Count[w][i]) + 1

  i=i+1
```

```python
#pd.set_option('display.max_columns', None)
Term_Frequency_Log_Normalisation
```

###Analysis
**Pros:** 
1. Dampen the effect of the term that has a high frequency.
2. Brings non-linearity in the scoring process.
3. We also add 1 to the log(tf) because when tf is equal to 1, the log(1) is zero. By adding one, we distinguish between tf=0 and tf=1.

**Cons:** 
1. For frequent terms, we want high +ve weights for the words, but  lower weights than the rare terms.
2. The method also does not capture the semantic correlation between different terms in a document.

```python
n_docs = len(list_dir)         #Number of documents in the corpus
n_words_set = len(vocab_list) #Number of unique words in the
 
Term_Frequency_Double_Normalisation = pd.DataFrame(np.zeros((n_docs, n_words_set)), columns=vocab_list)
 
#iterating through each file in directory
i=0
for file_name_ in tqdm(list_dir):
  file_name= file_path+file_name_

  with open(file_name, 'r') as f: #open the file in read mode

      #reading the file
      contents = f.read()
      
      #list containing each term in the file as an element
      words=list(contents.split(' '))
      for w in words:
        Term_Frequency_Double_Normalisation[w][i] = 0.5 + ( 0.5 * Term_Frequency_Raw_Count[w][i] / Term_Frequency_Raw_Count.loc[[i]].max(axis=1) )

  i=i+1
```

```python
#pd.set_option('display.max_columns', None)
Term_Frequency_Double_Normalisation
```

###Analysis
**Pros:** 
1. Also called "Augmented frequency", is used to prevent a bias towards longer documents, e.g. raw frequency divided by the raw frequency of the most frequently occurring term in the document.
2. Brings non-linearity in the scoring process.
3. It is therefore critical to balance between over-weighting and under-weighting.

**Cons:** 
1. Sublinear scaling and bias term shrink the ratios between weights of different terms. If the scaling functions scale down the weights too much or bias term is too large, ratios between term weights become too small. Hence under-weighting occurs.
2. Regularization techniques may be required to prevent under/over weighting, thus, adding to overhead.
